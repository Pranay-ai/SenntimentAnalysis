```python
import boto3

s3 = boto3.client('s3')

bucket_name = 'cpsc454sabucket'
subdirectory_name = 'tf-imdb-data/'


training_input_path=s3.upload_file('a1_IMDB_Dataset.csv', bucket_name, subdirectory_name + 'training/a1_IMDB_Dataset.csv')
```

```python

#Need tf version and role for the Tensorflow Estimator
tf_version = tf.__version__

#Sagemaker role, make sure you've allowed access to any S3 bucket
role = sagemaker.get_execution_role()
```

```python

training_input_path='s3://cpsc454sabucket/tf-imdb-data/training/a1_IMDB_Dataset.csv'
```

```python
from sagemaker.tensorflow import TensorFlow
tf_estimator = TensorFlow(entry_point='train_script.py', 
                          role=role,
                          instance_count=1, 
                          dependencies=['requirements.txt'],
                          instance_type='ml.m5.large',
                          framework_version=tf_version, 
                          py_version='py310',
                          script_mode=True,
                          hyperparameters={
                              'epochs': 3
                              
                          }
                         )

#Training
tf_estimator.fit({'train': training_input_path
                 })
```

```python
import time
tf_endpoint_name = 'tf-imdb-model'+time.strftime("%Y-%m-%d-%H-%M-%S", time.gmtime())
tf_predictor = tf_estimator.deploy(initial_instance_count=1,instance_type='ml.m5.large',
                                   endpoint_name=tf_endpoint_name)
print("Success!")
```
